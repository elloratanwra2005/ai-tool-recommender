```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import string

ps = PorterStemmer()

def preprocess(text):
    text = text.lower()
    words = text.split()

    words = [
        ps.stem(word)
        for word in words
        if word not in stopwords.words('english')
        and word not in string.punctuation
    ]
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
import re

new_df['tag'] = new_df['tag'].fillna('')

new_df['tag'] = new_df['tag'].apply(
    lambda x: re.sub(r'http\S+|www\S+', '', x)
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')

vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = tfidf.fit_transform(new_df['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(tool_name):
    tool_name = tool_name.strip()

    if tool_name not in new_df['Tool Name'].values:
        print("Tool not found!")
        return

    idx = new_df[new_df['Tool Name'] == tool_name].index[0]

    distances = similarity[idx]

    tool_list = sorted(
        list(enumerate(distances)),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    print(f"\nRecommendations for {tool_name}:\n")

    for i in tool_list:
        print(new_df.iloc[i[0]]['Tool Name'])
```

```python
recommend("Murf.ai")
```

```python
new_df = new_df.reset_index(drop=True)
```

```python
def recommend(tool_name):
    tool_name = tool_name.strip()

    matches = new_df[new_df['Tool Name'].str.lower() == tool_name.lower()]

    if matches.empty:
        print("Tool not found!")
        return

    idx = matches.index[0]

    distances = similarity[idx]

    recommended = sorted(
        list(enumerate(distances)),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    print(f"\nTop recommendations for '{tool_name}':\n")

    for i, score in recommended:
        print(f"{new_df.iloc[i]['Tool Name']} ({score:.3f})")
```

```python
recommend("Murf.ai")
```
